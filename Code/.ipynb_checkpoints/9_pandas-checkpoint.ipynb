# DataFrame

## List

```python
data1 = [['Bob', 1993, 173.5],
         ['Steve', 1996, 178.1],
         ['Jeff', 1994, 183.2]]
df1 = pd.DataFrame(data1,
                   index=['I', 'II', 'III'],
                   columns=['student', 'birthyear', 'height'])
df1
```

## Dict

```python
data2 = {'student': ['Bob', 'Steve', 'Jeff'],
         'birthyear': [1993, 1996, 1994],
         'height': [173.5, 178.1, 183.2]}
df2 = pd.DataFrame(data2,
                   index=['I', 'II', 'III'])
df2
```

# Missing Data

## drop

```python
df = df2.drop(columns=['名稱', '類型'])
df = df[df > 0]
df
```

## dropna

```python
df = df2.drop(columns=['名稱', '類型'])
df = df[df > 0]

df = df.dropna(axis=0,how='any')
df
```

```python
df = df2.drop(columns=['名稱', '類型'])
df = df[df > 0]

df = df.dropna(axis=1,how='any')
df
```

## fillna

```python
df = df2.drop(columns=['名稱', '類型'])
df = df[df > 0]

df = df.fillna(value=5)
df
```
